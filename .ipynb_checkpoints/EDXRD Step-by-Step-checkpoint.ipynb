## Data Directory

The data directory is broken down as follows. All text files and information are stored in a folder titled 'Spectra'. 'Spectra' has 3 sub-directories: 'Samples', 'Reference', and 'Calibration'. The data has already been scrubbed so that the directory only contains usable scans. For examples, 3 or 4 Mo W foil scans were recorded, but only 1 of them was suitable and a complete scan. The previous Mo W foil scans are not included in this data directory for simplicity of analysis. All original hdf files from the beamline are preserved.

### 'Calibration'
Stores all the spectra we took at the beamline that were NOT samples. This includes the Mo W foil, calcite powder, and LaB6 diffraction reference. 

### 'Reference'
Any diffraction pattern that didn't come from the October APS run at 7-BM. The format is a two columned text file, column one is 2-theta, column two is the intensity. 

### 'Samples'
Core of the operation. Within the samples subdirectory is a folder for each sample we brought to the beamline ('High pH BL 25C 4d', 'High pH SF Ref', 'Low pH SF 60C 4d'), and within these folders are the data for the single best spatial scan taken of that sample. Within a sample subdirectory are text files named numerically ('7.0', '12.25'). The text file '10.0' contains only a 1D column of intensity data, which is the EDXRD pattern taken 10.0 mm away from the sample's center.

The samples subdirectory includes two additional folders in each sample folder: 'energy' and 'background'. Energy contains a 1D column of the energy in keV of the associated bin, taken from my data fit of the Mo W foil reference. 'Background' contains a 1D column of the average off-sample background (Average of 3 scans taken off-sample) intensity. The EDXRD spatial data already has this off-sample background subtracted out.

```python
from pathlib import Path
import os

'''
READING FILES AND GETTING DATA INTO PYTHON
'''

#GET A LIST OF FILES IN A DIRECTORY WITH A GIVEN EXTENSION (i.e. 'txt')
def file_list(directory, ftype):
    allflist = sorted(os.listdir(directory))
    flistlen = len(allflist)
    flist = []
    i = 0
    for x in range(flistlen):
        if allflist[x].endswith(str(ftype)):
            # flist.append((allflist[x], i)); i +=1
            flist.append(allflist[x])
            
    if ('.DS_Store' in flist):
        flist.remove('.DS_Store')
    
    if ('.ipynb_checkpoints' in flist):
        flist.remove('.ipynb_checkpoints')
    
    return flist

#GIVE ME A ONE COLUMN TXT FILE AND RETURN AN ARRAY
def unpack_a_file(file):
    array = []
    with open(file, 'r') as f:
        for lines in f:
            data = float(lines.split('\n')[0])
            array.append(data)
    f.close()
    return array

#GIVE ME ONE SAMPLE AND RETURN THE SPECTRA AND EACH POSITION
def unpack_a_sample_dir(sample_dir):
    files = file_list(sample_dir, '.txt')
    scans = []

    energy = unpack_a_file(sample_dir / Path('energy/energy.txt'))
    background = unpack_a_file(sample_dir / Path('background/background.txt'))
        
    positions = []
    spectras = []
    
    for x in range(len(files)):
        pos = float(files[x].split('.txt')[0])
        positions.append(pos)
    positions.sort()
        
    for x in range(len(positions)):
        fname = Path((str(positions[x]) + '.txt'))
        spec = unpack_a_file(sample_dir / fname)
        spectras.append(spec)

    return energy, spectras, positions


#Define the sample directory location
sdir = Path('Spectra/Samples')

#Define the sample list
slist = file_list(sdir, '')

#Print the samples within the directory
slist
```

## Gaussian Fit to These Peaks
First of all -the data has not been adjusted for absorption. For later analysis, that will be very important. However, for now I am just trying to determine if the peak locations are consistent across our experiment. Based on the plots, Portlandite Peak 1 and Calcite Peaks 1 and 2 offer the most promise. The portlandite features are suppressed due to the lack of an absorption correction, but are still present in the data and usable for this purpose.

There are a few steps here:

### Step 1: Fit background to a polynomial and subtract fitted polynomial from spectrum

The reason I took the time to make the graphs above is because it shows the background vs peak regime. Everything enclosed between the two vertical lines is defined as the peak and not included in data for the fit. The remainder of data within the window is all to be fitted to a polynomial.

Once the polynomial is determined, subtracting it from each point should remove background and get baseline of the peaks and Gaussian fit on the x-axis (y=0)

### Step 2: Fit a gaussian to the defined peaks

Like before -the peak regime is whatever is in-between the two vertical lines in the plots above. That is the data that will be fitted to the gaussian

### Step 3: Average peak locations

I want to average these in two ways: Average them for the sample and compare each sample's peak location to another, and average them by spatial position and compare each peak location at a given beam position to another. Especially in the second check, I want to ensure there is not a trend or drift in all the peaks one way or another


```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from numpy.polynomial import polynomial as poly 
'''
FUNCTIONS TO ASSIST IN THE GAUSSIAN FIT AND POLYNOMIAL BACKGROUND REMOVAL
'''

#Definition of a gaussian function
# X - Value, H - Offset of Baseline, A - Amplitude, X0 - Center, Sigma - Standard Deviation
def gauss(x, H, A, x0, sigma):
    return H + A * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))

#Given an x-y data set, return H, A, x0, sigma fitting parameters
def gauss_fit(x, y):
    mean = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[min(y), max(y), mean, sigma])
    return popt

#Given a polynomials coeffecients, subtract the polynomial evaluated at x from ydata at x
def subtract_poly(xdata, ydata, coeff):
    if len(xdata) != len(ydata): return print('ERROR: Array dimension mismatch')
    y_adj = []
    pfit = poly.Polynomial(coeff)
    for g in range(len(xdata)):
        y_adj.append(ydata[g] - pfit(g))
    return y_adj
```

```python
#A Function that fits a gaussian to whatever window and range it gets given. This banks on the index 
#of the xdata being the same as the value, i.e. bins only! Don't use this with any other x-axis

def peak_gauss_fit(ydata, window, peak_bound):
    #get the background data fitted to a polynomial
    x1 = window[0]
    x2 = peak_bound[0]
    x3 = peak_bound[1]
    x4 = window[1]
    
    #Define x and y data outside the peak within the window to create polynomial coeffecients
    back_x = list(range(x1, x2+1)) + list(range(x3, x4+1))
    back_y = ydata[x1: x2+1] + ydata[x3: x4+1]
    pcoef = poly.polyfit(back_x, back_y, 1)
    
    #Create an array of the fit data
    poly_x = list(range(x1, x4+1))
    poly_y = []
    pfit = poly.Polynomial(pcoef)
    for x in range(len(poly_x)): poly_y.append(pfit(x))
    
    
    y_sub = subtract_poly(poly_x, ydata, pcoef)

    fig, ax = plt.subplots()
    ax.plot(poly_x, y_sub)
    
    
    return poly_x, poly_y, ydata
    
    
    
    
    
```
